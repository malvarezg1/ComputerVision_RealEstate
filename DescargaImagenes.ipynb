```python
def descargar(datos,start, end):
    global ruta
    data=datos
    global inmueble
    for inmueble in range(start,end):
        urls=data["imagenes"].iloc[inmueble]
        urlsList = urls[1:len(urls)-1].replace('"','').split(",")
        cont =0 
        for url in urlsList:
            cont = cont +1
            urllib.request.urlretrieve(url, ruta
                                       +data["tipoinmueble"].iloc[inmueble]+str(int(data["id"].iloc[inmueble]))+"_"+str(cont)+".jpg")
```

```python
ruta = "C:/Users/m.alvarezg1/Documents/DataU/"
descargar(filtrados,1025,1120)
```

```python
def descargarDataPrueba(datos,start, end):
    global ruta
    data=datos
    global inmueble
    dires=['bedroom','bathroom','kitchen','living_room','dining_room']
    for inmueble in range(start,end):
        urls=data["imagenes"].iloc[inmueble]
        urlsList = urls[1:len(urls)-1].replace('"','').split(",")
        cont =0 
        path = os.path.join(ruta, data["tipoinmueble"].iloc[inmueble]+"_"+str(int(data["id"].iloc[inmueble])))
        os.mkdir(path)
        for d in dires:
            path_d = os.path.join(path, d)
            os.mkdir(path_d)
        for url in urlsList:
            cont = cont +1
            urllib.request.urlretrieve(url, path+
                                       "/"+data["tipoinmueble"].iloc[inmueble]+"_"+str(int(data["id"].iloc[inmueble]))+"_"+str(cont)+".jpg")
```

```python
### Crear estructura de directorios
def crearEstrucutura(data):
    global ruta
    inmubles = data["inmueble_id"].unique().tolist()
    dires=['bedroom','bathroom','kitchen','living_room','dining_room']
    for inmu in inmubles:
        path=os.path.join(ruta,"a_"+str(inmu))
        os.mkdir(path)
        for d in dires:
            path_d = os.path.join(path, d)
            os.mkdir(path_d)
```

```python
ruta = "E:/TesisMA/DatosF2/"
crearEstrucutura(data_fase2)
```

```python
def descargarDataF2(data):
    global url
    global ruta
    for i in range(0,5):
        url_img = os.path.join(url,data["url"][i])
        path = os.path.join(ruta,"a_"+str(data["inmueble_id"][i]))
        path = os.path.join(path,str(i)+".jpeg")
        print(path)
        with open(path, 'wb') as handle:
            response=requests.get(url_img,headers={'Host': 'habi-sales-images-prod.s3.us-east-2.amazonaws.com','User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'})
            if not response.ok:
                print(response)

            for block in response.iter_content(1024):
                if not block:
                    break

                handle.write(block)
```

```python
ruta = "E:/TesisMA/DatosF2/"
url="http://habi-sales-images-prod.s3.us-east-2.amazonaws.com"
descargarDataF2(data_fase2)
```
