# Comparador de modelos

### Librerias e importaciones

```python
import numpy as np
import os
import PIL
import sklearn

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd
from datetime import datetime
from matplotlib import image
from sklearn.metrics import classification_report
```

```python
def test_props(model, data,img_height,img_width):
    #Setting de parametos
    props = []
    loss = []
    accuracy = []
    instances=[]
    statistics = {}
    os.chdir(data)
    properties = os.listdir()

    for p in properties:
        props.append(p)
        path_p = data+p
        #Crar dataset
        val_ds = tf.keras.preprocessing.image_dataset_from_directory(path_p, image_size=(img_height, img_width))
        instances.append(len(val_ds.file_paths))
        report = model.evaluate(val_ds,return_dict=True, verbose=0)
        loss.append(report['loss'])
        accuracy.append(report['accuracy'])
        statistics ["Propiedad"] = props
        statistics ["Instancias"] = instances
        statistics ["Loss"] = loss
        statistics ["Accuracy"] = accuracy
    return pd.DataFrame(statistics) 
```

```python
def comparate_models(data_root, path_models, img_height, img_width):
    os.chdir(path_models)
    models = os.listdir()
    acc = []
    accW= []
    sd =[]
    for m in models:
        path_m = path_models+"/"+m
        model = tf.keras.models.load_model(path_m)
        results=test_props(model, data_root, img_height, img_width)
        acc.append(results["Accuracy"].mean())
        accW.append((results["Accuracy"] * results["Instancias"]).sum() / results["Instancias"].sum()) 
        sd.append(results["Accuracy"].std()) 
    comp = {}
    comp["Modelos"]=models
    comp["Accuracy Instancias"] =accW
    comp["Mean Accuracy Properties"]=acc
    comp["Accuracy SD"]=sd
    return pd.DataFrame(comp)
```

```python
data_root = "E:/TesisMA/Datos/DataPrueba/"
path_models = 'E:/TesisMA/Modelos/512'

img_height=512
img_width=512
comparate_models(data_root, path_models, img_height, img_width)
```
