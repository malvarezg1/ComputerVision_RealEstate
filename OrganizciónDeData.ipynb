```python
import os
import shutil
import pandas as pd
```

```python
path="C:/Users/m.alvarezg1/Downloads/DataT.csv"
df=pd.read_csv(path,header=None)
df
```

```python
df[1]=df[1].str.extract(r'\[(.*?)\]')
df.loc[df[1]=='living room',1]='living_room'
df.loc[df[1]=='dining room',1]='dining_room'
df
```

```python
pathOrigin="C:/Users/m.alvarezg1/Documents/DataT/"
pathDestination="E:/TesisMA/Datos/DataCombinada/"
for i in range(0,df.shape[0]):
    tempPath=pathOrigin+df[0].iloc[i]
    tempPath1=pathDestination+df[1].iloc[i]
    shutil.copy2(tempPath, tempPath1)
    os.remove(pathOrigin+df[0].iloc[i])
```
